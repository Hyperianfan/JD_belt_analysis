### 华为手环差评分析

```python
import pandas as pd
df_comment_huawei = pd.read_csv("comments_chaping_huawei_B5_999.csv","utf-8",header=None,names=["content"])
df_comment_huawei = df_comment_huawei.drop_duplicates()
df_comment_huawei.head()
```

```python
df_comment_huawei["time"] = df_comment_huawei["content"].str.findall(r'(\d+-\d+-\d+ \d+:\d+:\d+)')
df_comment_huawei.head()
```

```python
df_comment_huawei["contents"] = df_comment_huawei["content"].str.findall(r'[\u4e00-\u9fa5]+')
df_comment_huawei = df_comment_huawei.drop(["content"],axis =1)
df_comment_huawei.head()
```

```python
df_comment_huawei.to_csv("huawei.csv",index=False)
```

```python
str_data = ''
for i in range(df_comment_huawei.shape[0]):
    str_data = str_data + str(df_comment_huawei.iloc[i,1])
str_data
```

```python
import jieba.analyse
tags = jieba.analyse.extract_tags(str_data, topK=50, withWeight=True)
label = []
attr = []
for item in tags:
    print(item[0]+'\t'+str(int(item[1]*1000)))
    label.append(item[0])
    attr.append(int(item[1]*1000))
```

```python
words = [list(z) for z in zip(label,attr)]

from pyecharts import options as opts
from pyecharts.charts import Page, WordCloud
from pyecharts.globals import SymbolType

c = (
    WordCloud()
    .add("", words, word_size_range=[20, 100])
    .set_global_opts(title_opts=opts.TitleOpts(title="华为手环差评词云"))
)
c.render("词云.html")
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar

c = (
    Bar()
    .add_xaxis(["重量轻","包装难看","蓝牙通话延迟","信息泄露","安装软件过多","外面瑕疵","自动重启","充不进电",
              "蓝牙自动断开","无适配器，数据线","续航不足","通话有杂音","屏幕易碎","手腕上不能通话","天气温度不准",
               "睡眠监测不准","不能收微信信息","耳套容易掉",])
    .add_yaxis("差评数目统计（选取100人为样本）", [1,1,1,1,2,2,2,3,3,4,5,6,6,7,7,8,8,10])
    
    .reversal_axis()
    .set_series_opts(label_opts=opts.LabelOpts(position="right"))
    .set_global_opts(title_opts=opts.TitleOpts(title="华为差评最多项"))
)
c.render("华为差评最多项barh.html")
```
