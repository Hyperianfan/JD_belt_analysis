```python
JD_Belt_Func = pd.read_csv("JD_shouhuan_with_function_new.csv",header=None, names=["brand","contents"] )

JD_Belt_Func.drop_duplicates()
JD_Belt_Func.shape
```

```python
JD_Belt_Func["contents"].str.findall(r'(?<=商品毛重)：(\d+\.\d+|\d+)')
```

```python
JD_Belt_Func["p_name"] = JD_Belt_Func["contents"].str.findall(r'(?<=商品名称)：(\w+)')
JD_Belt_Func["sku_id"] = JD_Belt_Func["contents"].str.findall(r'(?<=商品编号)：(\d+)')
JD_Belt_Func["weight"] = JD_Belt_Func["contents"].str.findall(r'(?<=商品毛重)：(\d+\.\d+|\d+)')
JD_Belt_Func["local"] = JD_Belt_Func["contents"].str.findall(r'(?<=商品产地)：(\w+)')
JD_Belt_Func["control"] = JD_Belt_Func["contents"].str.findall(r'(?<=触控方式)：(\w+)')
JD_Belt_Func["inch"] = JD_Belt_Func["contents"].str.findall(r'(?<=屏幕尺寸)：(\d+.\d+-\d+\.\d+|\d+\.\d+)')
JD_Belt_Func["connect"] = JD_Belt_Func["contents"].str.findall(r'(?<=连接方式)：(\w+)')
JD_Belt_Func["communicate"] = JD_Belt_Func["contents"].str.findall(r'(?<=通讯功能)：(\w+)')
JD_Belt_Func["materal"] = JD_Belt_Func["contents"].str.findall(r'(?<=腕带材质)：(\w+)')
JD_Belt_Func["IPXX"] = JD_Belt_Func["contents"].str.findall(r'(?<=防水等级)：(\w+)')
JD_Belt_Func["xuhang"] = JD_Belt_Func["contents"].str.findall(r'(?<=续航时间)：(\d+-\d+|\d+)')
JD_Belt_Func["display"] = JD_Belt_Func["contents"].str.findall(r'(?<=屏幕显示)：(\w+)')
JD_Belt_Func["funy"] = JD_Belt_Func["contents"].str.findall(r'(?<=运动模式识别)：(\w+)')
JD_Belt_Func["funy"] = JD_Belt_Func["contents"].str.findall(r'(?<=运动模式识别)：(\w+)')
JD_Belt_Func["severce"] = JD_Belt_Func["contents"].str.findall(r'(?<=优选服务)：(\w+)')
JD_Belt_Func["is_bluetooth_talk"] = JD_Belt_Func["contents"].str.findall(r'(?<=蓝牙通话)：(\w+)')
JD_Belt_Func["people"] = JD_Belt_Func["contents"].str.findall(r'(?<=适用人群)：(\w+)')
JD_Belt_Func["function"] = JD_Belt_Func["contents"].str.findall(r'(?<=功能用途)：(\w+)')


```
